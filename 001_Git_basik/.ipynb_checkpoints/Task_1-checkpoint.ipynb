```python
first_phrase = 'Насколько проще было бы писать программы, если бы не заказчики'
second_phrase = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
if first_phrase > second_phrase:
    print('Фраза 1 длиннее фразы 2')
elif first_phrase == second_phrase:
    print ('Фразы равной  длины')
else: 
    print ('Фраза 2 длиннее фразы 1')
```
