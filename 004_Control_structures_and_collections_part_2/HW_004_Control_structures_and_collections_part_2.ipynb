## Задание 1

```python
ids = {'user1': [213, 213, 213, 15, 213],
       'user2': [54, 54, 119, 119, 119],
       'user3': [213, 98, 98, 35]}
# Создаем пустой список в который будут переданы все значения из словаря ids
ids_val = []
# Создание отдельного двумерного списка значений словаря
ids_all_values = list(ids.values())
# Преобразование двумерного списка в одномерный с помощью 2-х вложенных циклов for
for elements in range(len(ids_all_values)):    
    for inner_elements in range(len(ids_all_values[elements])):
        ids_val.append(ids_all_values[elements][inner_elements])
print('Результат: ', set(ids_val))
```

## Задание 2

```python
queries = [
'смотреть сериалы онлайн',
'новости спорта',
'афиша кино',
'курс доллара',
'сериалы этим летом',
'курс по питону',
'сериалы про спорт',
]
words_quantity =[]
# Определяем количество слов в каждом пункте списка queries
# Разбиваем пункты списка на слова по разделителю () с помощью метода split
for item in queries:
    words_quantity.append(len(item.split()))
#Расчет процентного соотношения количества слов в запросах
#Вводим переменную для содержащую общее количество пунктов списка
total_reqv = len(queries)
#Расчет доли запросов с 2 / 3 словами
two_words = words_quantity.count(2)/ total_reqv * 100
three_words = words_quantity.count(3)/ total_reqv * 100
#Расчет и вывод ответа
print('Поисковых запросов, содержащих 2 слов(а):',round(two_words,2),'%')
print('Поисковых запросов, содержащих 2 слов(а):',round(three_words,2),'%')
```
