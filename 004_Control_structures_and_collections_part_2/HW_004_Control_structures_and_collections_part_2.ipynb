## Задание 1

```python
ids = {'user1': [213, 213, 213, 15, 213],
       'user2': [54, 54, 119, 119, 119],
       'user3': [213, 98, 98, 35]}
# Создаем пустой список в который будут переданы все значения из словаря ids
ids_val = []
# Создание отдельного двумерного списка значений словаря
ids_all_values = list(ids.values())
# Преобразование двумерного списка в одномерный с помощью 2-х вложенных циклов for
for elements in range(len(ids_all_values)):    
    for inner_elements in range(len(ids_all_values[elements])):
        ids_val.append(ids_all_values[elements][inner_elements])
print('Результат: ', set(ids_val))
```
