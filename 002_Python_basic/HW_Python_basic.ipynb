## Задание 4

```python
weight = int(input('Введите ширину коробки в см: '))
length = int(input('Введите длину коробки в см: '))
height = int(input('Введите высоту коробки в см: '))
if length > 200:
    print('Упаковка для лыж')
elif weight < 15 and length < 15 and height < 15:
    print('Коробка №1')
elif 15 <= weight <= 50 or 15 <= length <= 50 or 15 <= height <= 50:
    print('Коробка №2')
else:
    print('Стандартная коробка №3')
```
