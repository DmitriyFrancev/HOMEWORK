## Задание 1

```python
test_word = input('Введите слово: ')
letter_half = int(len(test_word)/2)           # Рассчитываем половину количества букв в слове
if (len(test_word)) % 2 == 0:                  # Проверка, является ли число букв четным?
    print(test_word[letter_half - 1] + test_word[letter_half])
else:
    print (test_word[letter_half])
    
```

## Задание 2

```python
user_number = int(input('Введите число: '))
total = 0
while user_number != 0:
    total += user_number
    user_number = int(input('Введите число: '))
else:
    print('Результат: ', total)
    
```

## Задание 3

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
if len(boys) == len(girls):
    boys.sort()
    girls.sort()
    result = [x + ' и ' + y for (x,y) in zip(boys, girls)]
    print('Результат: ')
    print('Идеальные пары: ')
    for pairs in result:
        print(pairs)
else:
    print('''Результат:
    Внимание, кто-то может остаться без пары!''')    
```

## Задание 4

```python
countries_temperature = [
['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
['Germany', [57.2, 55.4, 59, 59, 53.6]],
['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
print('Средняя температура в странах: ')
for country, temperature, in countries_temperature:
    result_celcium = (sum(temperature) / len(temperature) - 32) * 5 / 9
    print(country, '-', round(result_celcium, 1), 'C')
```
